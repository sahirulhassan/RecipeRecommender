```python
import pandas as pd
import re
```

```python
recipes = pd.read_csv("C:/Users/sahir/Documents/Personal/Usman Institute of Technology University/Semester III/DSA/recipe_Recommender.RecipeRecommender/src/test/resources/dataset_cleaned.csv")
recipes
```

```python
target_col = ['ingredients', 'ingredients_raw', 'steps']
target_col2 = ['tags']
```

```python
def cleaner(input):
    if isinstance(input, str):
        output = re.sub(r'\s+', ' ', input) #for unnecessary whitespaces
        output = output.replace('["', '') #for opening json brackets
        output = output.replace('","', '\n') #for line breaks
        output = output.replace('", "', '\n') #for line breaks
        output = output.replace('" ,"', '\n') #for line breaks
        output = output.replace('"]', '\n\n') #for paragraph breaks
        return output

def stringifier(df, target_cols):
    for column in target_cols:
        df[column] = df[column].apply(cleaner)
    return df

#driver code
new_dataset = stringifier(recipes, target_col2)
new_dataset.to_csv("C:/Users/sahir/Documents/Personal/Usman Institute of Technology University/Semester III/DSA/recipe_Recommender.RecipeRecommender/src/main/resources/dataset_cleaned_v2.csv", index=False)
new_dataset
```
